###Find the sum of all multiples of n below m
-n and m are natural numbers (positive integers)

-m is excluded from the multiples

```python
def sum_mul(n, m):
    x=False
    if (n>0 and m >0):
        if (n<m):
            for i in range(n,m):
                if i%n==0:
                    x+=i
            return x
        elif n >= m:
            return 0
    else:
        return 'INVALID'
sum_mul(10,80)
```
