###Given a year, return the century it is in.

```python
def century(year):
    a=str(year)
    if len(a)==4:
        if (year / 100 == int(a[0:2])):
            return(int(a[0:2]))
        else:
            return(int(a[0:2])+1)
    elif len(a)==3:
        if (year % 100 ==0):
            return(int(a[0:1]))
        else:
            return(int(a[0:1])+1)
    elif len(a)<=2:
        return(1)

century(154)
```
