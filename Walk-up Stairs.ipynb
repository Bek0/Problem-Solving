###Raj was to move up through a pattern of stairs of a given number (n). Help him to get to the top using the function stairs.

###Keep in mind :

###If n<1 then return ' ' .
###There are a lot of spaces before the stair starts except for pattern(1)

```python
def stairs(n):
    if n >= 1:
        z=n*2+1
        x=n
        v=0
        a=1
        w=1
        list3=[]
        for i in range(0,n):
            list1=[]
            list2=[]
            for p in range(1,z):
                if x:
                  list1.append(int(str(p)[-1]))
                  x=x-1
                else:
                    x=n
                    x=x-a
                    a+=1
                    list1.append(list1[::-1])
                    if v == 0:
                        v=len(list1[-1])

                    list2=list1[-1]
                    del list1[-1]
                    list3.append(list1+list2)
                    list3.append("\n")
                    w=5
                    break
        del list3[-1]
        r = (v * 2) + ((v * 2) - 1) - 3
        ok = ""
        for i in list3[::-1]:
            if type(i)==list:
                ok += " " * (r)
                r -= 4
                for k in range(len(i)):
                    if (k == len(i)-1):
                        ok+= f"{i[k]}"
                    else:
                        ok+= f"{i[k]} "
            else:
                ok+='\n'
        return ok
    else:
        return ''


print(stairs(8))
print(stairs(15))
```
