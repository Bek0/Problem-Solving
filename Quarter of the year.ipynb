###Given a month as an integer from 1 to 12, return to which quarter of the year it belongs as an integer number.


```python
def quarter_of(month):
    if month<=12 :
        x=month/3
        if x<=1:
            return 1
        elif x<=2:
            return 2
        elif x<=3:
            return 3
        elif x<=4:
            return 4
    else:
        while(month > 12):
            month-=12
        return quarter_of(month)

quarter_of(8)
```
